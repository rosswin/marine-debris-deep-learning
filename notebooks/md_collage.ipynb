```python
import os
import math
from PIL import Image
```

```python
gt_path = r"/home/ross/Documents/20201025_final_results/groundtruth_png"
ir_path = r"/home/ross/Documents/20201025_final_results/incept_resnet/results/png"
mn_path = r"/home/ross/Documents/20201025_final_results/mobilenet/results/png"
out_path = r"/home/ross/Documents/20201025_final_results/collage2021"
```

```python
gt_imgs = set([f for f in os.listdir(gt_path)])
ir_imgs = set([f for f in os.listdir(ir_path)])
mn_imgs = set([f for f in os.listdir(mn_path)])
```

```python
intersect = set.intersection(gt_imgs, ir_imgs, mn_imgs)
```

```python
print(len(gt_imgs), len(ir_imgs), len(mn_imgs), len(intersect))
```

```python
for img in list(intersect)[:1]:
    print(img)
```

```python
for img in list(intersect):
    gt_in = os.path.join(gt_path, img)
    gt_im = Image.open(gt_in)
    ir_in = os.path.join(ir_path, img)
    ir_im = Image.open(ir_in)
    mn_in = os.path.join(mn_path, img)
    mn_im = Image.open(mn_in)

    total_width = gt_im.size[0] + ir_im.size[0] + mn_im.size[0] + 150
    total_height = max(gt_im.size[1], ir_im.size[1], mn_im.size[1])
    print(total_width, total_height)
    print(gt_im.size, ir_im.size, mn_im.size)

    new_img = Image.new('RGB', (total_width, total_height), 'white')
    new_img.paste(ir_im, (0,0))
    new_img.paste(mn_im, (ir_im.size[0] + 50, 0))
    new_img.paste(gt_im, (ir_im.size[0] + mn_im.size[0] + 100,0))
    
    new_img.save(os.path.join(out_path, img))
```
